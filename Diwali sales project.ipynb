# Exploratory Data Analysis(EDA)

```python
df.columns
```

```python
#To find which gender is shopping more
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers: 
    ax.bar_label(bars) # to put labels on top of the graph
```

### Group by, order by and sum in Python

```python
sales_value=df.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False) 
```

```python
sns.barplot(x='Gender', y='Amount', data= sales_value)
```

* From the above graphs we can say that the number of female buyers are more
than the male buyers and female buyers have more purchasing power as compared to male buyers

### Analysis on the basis of age group

```python
ax= sns.countplot(data=df,x='Age Group', hue='Gender')  # numbers of sales by age group
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# amount of sales by age group
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount',data=sales_age)
```

```python
# Group by states and sorting by orders top 10 states
```

```python
sales_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State', y='Orders')
```

```python
sales_state_amt=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state_amt,x='State', y='Amount')
```

### Analysis on the basis of marital status

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(3,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_ms,x='Marital_Status',y='Amount',hue='Gender')
```
