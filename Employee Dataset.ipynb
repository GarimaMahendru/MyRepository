# Scatter plot of age and employee salary

```python
plt.figure(figsize=(9,4))
sns.scatterplot(data=data, x='age', y='salary', color='red',edgecolor='linen',alpha=0.7)
plt.title("scatter plot for age and salary")
plt.xlabel("age")
plt.ylabel("salary")
plt.show()
```

# Scatter plot for healthy eating and active lifestyle

```python
plt.figure(figsize=(9,4))
sns.scatterplot(data=data, x='healthy_eating',y='active_lifestyle', color='blue',edgecolor='linen',alpha=0.7)
plt.title("scatter plot for healthy eating and active lifestyle")
plt.xlabel("healthy eating")
plt.ylabel("active lifestyle")
plt.show()
```

# Scatter plot between healthy eating and salary

```python
plt.figure(figsize=(9,4))
sns.scatterplot(data=data, x='healthy_eating',y='salary',color='pink',edgecolor='linen',alpha=0.7)
plt.title('Scatter plot of healthy eating and salary')
plt.xlabel('Healthy eating')
plt.ylabel('Salary')
plt.show()
```

# Count plot of groups

```python
plt.figure(figsize=(9,4))
sns.countplot(data=data, x='groups', edgecolor='linen',alpha=0.7)
plt.title('countplot of groups')
plt.xlabel('groups')
plt.ylabel('count')
plt.show()
```

# Histogram of Salary

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data, x='salary',edgecolor='linen',alpha=0.7,bins=10)
plt.title("histogram of salary")
plt.xlabel('salary')
plt.ylabel('amount')
plt.show()
```
